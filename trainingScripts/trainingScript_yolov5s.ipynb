### Train a Small (yolov5s) Model

Training all the layers of the small model.

yolov5s is a small model that is good when you want to run the model on a CPU and you don't have a lot of complexity. The pretrained weights associated with that model are our starting point. These pretrained weight have been created by training yolov5 on a completely different (publically available) dataset which was very large -> when starting from a pretrained model we don't need that many epochs to get good results.

we set the image size for training to 640. could also use 320 or something different

## show training progress in Tensorboard
https://www.tensorflow.org/tensorboard/tensorboard_in_notebooks

```python
from tensorboard import notebook
import os
import tempfile

def monitorTensorboard():
    # List open TensorBoard instances
    notebook.list()

    # Get the temporary directory path
    temp_dir = os.path.join(tempfile.gettempdir(), ".tensorboard-info")

    # List the files in the temporary directory which equals the size of the notebook list
    file_list = os.listdir(temp_dir)
    num_files = len(file_list)

    # Check if there are any instances
    if num_files > 0:
        print("Reloading TensorBoard...")
        %reload_ext tensorboard
        %tensorboard --logdir ../training_results/runs/train/
        notebook.display() #default port 6006 is chosen
    else:
        print("Starting TensorBoard...")
        %load_ext tensorboard
        %tensorboard --logdir ../training_results/runs/train/
```

```python
monitorTensorboard()
```
